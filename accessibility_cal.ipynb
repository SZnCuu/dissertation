# Single

```python
d_0 = 60
```

```python
gdf_access['single_step1'] = gdf_access['elderly_pop']*(1/gdf_access['time'])
```

```python
gdf_access.head()
```

```python
def single(gdf, group_col, filter_col, sum_col, d_0):
    def sum_single(df):
        filtered_df = df[df[filter_col] < d_0]

        if not filtered_df.empty:
            return filtered_df[sum_col].sum()
        return 0
    
    result = gdf.groupby(group_col).apply(sum_single).reset_index(name='sum_result')
    return result
```

```python
single_step1_result = single(gdf_access, 'destination', 'time', 'single_step1',d_0)
```

```python
single_step1_result = single_step1_result.rename(columns={'sum_result': 'single_step1_result'})
single_step1_result.head()
```

```python
# gdf_access = gdf_access.drop(columns=['single_step1_result_y', 'single_step1_result_x'])
```

```python
gdf_access = gdf_access.merge(single_step1_result, on='destination')
```

```python
gdf_access.head()
```

```python
gdf_access['single_step2'] = gdf_access['bed']*gdf_access['elderly_pop']/(gdf_access['time']*gdf_access['single_step1_result'])
```

```python
gdf_access.head()
```

```python
single_result = single(gdf_access, 'origin', 'time', 'single_step2',d_0)
```

```python
single_result = single_result.rename(columns={'sum_result': 'single_result'})
single_result
```

```python
sorted_single_result = single_result.sort_values(by='single_result', ascending=False)
sorted_single_result
```

# Classification - 3

```python
def time_class3(bed):
    if bed <= 100:
        return 30
    elif 100 < bed <= 300:
        return 60
    else:
        return 120
```

```python
gdf_access['time_class3'] = gdf_access['bed'].apply(time_class3)
```

```python
gdf_access.sample(n=5)
```

```python
def class_time(gdf, group_col, filter_col, sum_col, threshold_col):
    def sum_class(df):
        filtered_df = df[df[filter_col] < df[threshold_col]]
        
        if not filtered_df.empty:
            return filtered_df[sum_col].sum()
        return 0

    result = gdf.groupby(group_col).apply(sum_class).reset_index(name='sum_result')
    return result
```

```python
class3_step1_result = class_time(gdf_access, 'destination', 'time', 'single_step1','time_class3')
```

```python
class3_step1_result = class3_step1_result.rename(columns={'sum_result': 'class3_step1_result'})
class3_step1_result.head()
```

```python
# gdf_access = gdf_access.drop(columns=['class3_step1_result'])
```

```python
gdf_access = gdf_access.merge(class3_step1_result, on='destination')
```

```python
gdf_access['class3_step2'] = gdf_access['bed']*gdf_access['elderly_pop']/(gdf_access['time']*gdf_access['class3_step1_result'])
```

```python
class3_result = class_time(gdf_access, 'origin', 'time', 'class3_step2','time_class3')
```

```python
class3_result = class3_result.rename(columns={'sum_result': 'class3_result'})
class3_result
```
